```python
def str_ascii(str): 
    sum = 0
    ascii_values = [ord(c) for c in str]
    for x in ascii_values:
        sum = sum + x
    return sum
```

```python
def binary_search(arr, low, high, word):
    if high >= low: 
        mid = (high + low)/2
        if arr[mid] == word: 
            return mid
        
        elif arr[mid] > word:
            return binary_search(arr, low, mid-1, word)

        else: 
            return binary_search(arr, mid+1, high, word)
    else:
        return -1
```

```python
def get_character_frequencies(): 
    possible_matches = {}
    for word in words: 
        matches = get_character_matches(word)
        if len(matches) > 0: 
            possible_matches[word] = len(matches)

    possible_matches = sorted(possible_matches.items(), key=lambda x: x[1], reverse=True)
    possible_matches = [i[0] for i in possible_matches]
    return possible_matches
```
